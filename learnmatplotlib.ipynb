```python
#Tgnr hp tt c cc bi
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams["font.family"] = "Arial"

x = np.linspace(0, 2*np.pi, 300)
y = np.sin(x)
y1 = np.cos(x)

fig, ax = plt.subplots(figsize=(12,6))                     # resize figure
ax.set_facecolor("#888888")                                # background color

ax.plot(x, y, linestyle="--", linewidth=2, marker="o",    # markers + styling
        markersize=3, label="sin(x)")
ax.plot(x, y1, linestyle="-.", linewidth=2, marker="s",
        markersize=3, alpha=0.75, label="cos(x)")          # trans

ax.grid(True, which="both", linestyle="--", linewidth=0.8, alpha=0.6) # grid

ax.set_title("Matplotlib", fontsize=16)            # label
ax.set_xlabel("Truc X", fontsize=12, labelpad=18)
ax.set_ylabel("Truc Y", fontsize=12, labelpad=18)

ax.xaxis.set_label_coords(0.5, -0.10)                     # di chuyn cc label cnh
ax.yaxis.set_label_coords(-0.08, 0.5)

ax.spines["top"].set_visible(False)                       # Xa bolder
ax.spines["right"].set_visible(False)

ax.legend(frameon=False)
fig.tight_layout()
plt.show()
```
