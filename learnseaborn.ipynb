```python
#boxplot và violinplot
sns.set_theme()
tips = sns.load_dataset("tips").rename(columns={"day": "Ngày", "total_bill": "bill"})
fig, axes = plt.subplots(1, 2, figsize=(10,4))

sns.boxplot(data=tips, x="Ngày", y="bill", ax=axes[0])
axes[0].set_title("Boxplot")

sns.violinplot(data=tips, x="Ngày", y="bill", ax=axes[1])
axes[1].set_title("Violinplot")

plt.show()
```

```python
data = sns.load_dataset("titanic")

sns.set_theme()
sns.swarmplot(x='class', y='age', hue='sex', data=data)
plt.title("Phan bo tuoi hanh khach Titanic theo hang va gioi tinh")
plt.show()
```

```python
#catplot
data = sns.load_dataset("titanic")

sns.catplot(
    x="class", 
    y="age", 
    hue="sex", 
    kind="box", 
    data=data,
    height=5, 
    aspect=1.2
)

plt.title("Phan bo tuoi hanh khach theo hang va gioi tinh (box plot)")
plt.show()
```

```python
#displot
data = sns.load_dataset("titanic")

sns.displot(data["age"], bins=30, kde=False, color="skyblue")
plt.title("Phan phoi do tuoi hanh khach Titanic (Histogram)")
plt.xlabel("Tuoi")
plt.ylabel("Tan suat")
plt.show()
```

```python
#jointplot
data = sns.load_dataset("titanic")

sns.jointplot(x="age", y="fare", data=data)
plt.suptitle("Moi quan he giua tuoi va gia ve (Scatter + Hist)", y=1.02)
plt.show()
```

```python
#pairplot
data = sns.load_dataset("titanic")

sns.pairplot(data[["age", "fare", "pclass", "survived"]])
plt.suptitle("Moi quan he giua tuoi, gia ve, hang va trang thai song sot", y=1.02)
plt.show()
```

```python
#rugplot và KDE plot
data = sns.load_dataset("iris")

sns.kdeplot(x=data["sepal_length"], fill=True, color="skyblue")
sns.rugplot(x=data["sepal_length"], color="black")
plt.title("KDE + Rugplot: mat do va vi tri du lieu")
plt.xlabel("Sepal length")
plt.show()
```

```python
sns.kdeplot(data=data, x="sepal_length", hue="species", fill=True)
sns.rugplot(data=data, x="sepal_length", hue="species")
plt.title("Phan bo sepal_length theo tung loai hoa")
plt.show()
```

```python
#KDE plot
data = sns.load_dataset("iris")

sns.kdeplot(data=data, x="sepal_length", y="sepal_width", fill=True, cmap="mako")
plt.title("KDE 2 chieu")
plt.show()
```

```python
#Implot với regplot tương tự nhau
data = sns.load_dataset("tips")

sns.lmplot(x="total_bill", y="tip", data=data)
plt.title("")
plt.show()
```
